# Elden Ring - Weapons

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import ast
```

```python
weapons_df = pd.read_csv('weapons.csv')
```

Este dataset se encuentra disponible en: https://www.kaggle.com/datasets/robikscube/elden-ring-ultimate-dataset

```python
weapons_df.info()
```

Con la información que arroja la función `info()` sabemos que el dataframe se encuentra compuesto por 10 columnas: 1 de tipo float64 y 9 de tipo object, y no tiene valores nulos.

```python
weapons_df
```

Ahora que podemos ver más características del dataframe descubrimos que los registros de las columnas `attack`, `defence`, `scalesWith` y `requiredAttributes` son listas de diccionarios pero en formato object, por lo tanto, primero hay que convertir las columnas en listas para luego poder extraer su contenido y así poder analizar correctamente el dataframe en su totalidad.

Las columnas a extraer son `attack` y `requiredAttributes`, las columnas `description`, `defence` y `scalesWith` no serán tomadas en cuenta ya que no aportan valor al resultado final del análisis que se busca realizar.

```python
weapons_df = weapons_df.drop(['description', 'defence', 'scalesWith'], axis=1)
```

## Extracción de la columna `attack`

```python
# Convertir los registros de la columna attack a minusculas para evitar discrepancias
weapons_df['attack'] = weapons_df['attack'].str.lower()

# Convertir registros de la columna attack en listas
weapons_df['attack'] = weapons_df['attack'].apply(ast.literal_eval)

# Extraer las listas en un nuevo dataframe
attack_df = weapons_df['attack'].apply(pd.Series)

attack_df.info()
```

Este nuevo dataframe está formado por los diccionarios almacenados anteriormente en listas por la columna `attack`, cuenta con 7 columnas donde solo la última de ellas contiene valores nulos.

```python
attack_df[attack_df[6].notnull()]
```

Podemos ver que cada columna corresponde a un tipo de ataque, sin embargo la última columna no tiene un nombre fijo, variando entre Rng, Sor, Inc y Sorc. El siguiente paso es extraer el valor de `amount` de cada diccionario en una columna dedicada a cada tipo de ataque en el dataframe `weapons_df`.

Las columnas a insertar en el dataframe `weapons_df` son: `phyAttack`, `magAttack`, `fireAttack`, `lightAttack` y `holyAttack`. Los valores de última columna del nuevo dataframe no serán insertados en `weapons_df` ya que contiene demasiados valores nulos.

```python
damage_types_list = ['phyAttack', 'magAttack', 'fireAttack', 'lightningAttack', 'holyAttack', 'critAttack']

for i, damage_type in enumerate(damage_types_list):
    df = attack_df[i].apply(pd.Series)
    weapons_df[damage_type] = df['amount']

del attack_df

weapons_df.head()
```

## Extracción de la columna `requiredAttributes`

```python
# Convertir los registros de la columna attack a minusculas para evitar discrepancias
weapons_df['requiredAttributes'] = weapons_df['requiredAttributes'].str.lower()

# Convertir registros de la columna requiredAttributes en listas
weapons_df['requiredAttributes'] = weapons_df['requiredAttributes'].apply(ast.literal_eval)

# Extraer las listas en un nuevo dataframe
required_df = weapons_df['requiredAttributes'].apply(pd.Series)

required_df.info()
```

Este nuevo dataframe está formado por los diccionarios almacenados anteriormente en listas por la columna `requiredAttributes`, cuenta con 5 columnas donde solo la primera de ellas no contiene valores nulos.

```python
required_df
```

```python
required_df[(required_df[2].notnull())]
```

```python
required_df[(required_df[3].notnull())]
```

Se observa que la distribución de los diccionarios es similar a la obtenida anteriormente en la extracción de la columna `attack` excepto por la `columna 2` y la `columna 3`; ya que existen 4 registros donde la `columna 2` contiene solo los atributos y la `columna 3` contiene solo los valores correspondientes a los atributos de la `columna 2`.

```python
for i in range(5):
    df = required_df[i].apply(pd.Series)
    print(f"Col {i}: {df['name'].unique()}")
```

Una vez que conocemos los atributos presentes en cada columna podemos proseguir con la creación de las nuevas columnas correspondientes a cada tipo de atributo requerido.

Las columnas a insertar en el dataframe `weapons_df` son: `strRequired`, `dexRequired`, `intRequired`, `faiRequired` y `arcRequired`.

```python
# Columna 0
df = required_df[0].apply(pd.Series)
required_df['strRequired'] = df.loc[df['name'] == 'str', 'amount']
required_df['faiRequired'] = df.loc[df['name'] == 'fai', 'amount']
required_df['intRequired'] = df.loc[df['name'] == 'int', 'amount']

# Columna 1
df = required_df[1].apply(pd.Series)
required_df['dexRequired'] = df.loc[df['name'] == 'dex', 'amount']
required_df['intRequired'].fillna(df.loc[df['name'] == 'int', 'amount'], inplace=True)
required_df['faiRequired'].fillna(df.loc[df['name'] == 'fai', 'amount'], inplace=True)
required_df['arcRequired'] = df.loc[df['name'] == 'arc', 'amount']

# Columna 2
df = required_df[2].apply(pd.Series)
required_df['arcRequired'].fillna(df.loc[df['name'] == 'arc', 'amount'], inplace=True)
required_df['intRequired'].fillna(df.loc[df['name'] == 'int', 'amount'], inplace=True)
required_df['faiRequired'].fillna(df.loc[df['name'] == 'fai', 'amount'], inplace=True)
required_df['dexRequired'].fillna(df.loc[df['name'] == 'dex', 'amount'], inplace=True)

# Columna 3
df = required_df[3].apply(pd.Series)
required_df.loc[[146, 150, 257, 260]] = required_df.loc[[146, 150, 257, 260]].replace(0.0, 1)

# Columna 4
df = required_df[4].apply(pd.Series)
required_df['faiRequired'].fillna(df.loc[df['name'] == 'fai', 'amount'], inplace=True)

required_df = required_df.drop([0, 1, 2, 3, 4], axis=1)

# Reordenar columnas
required_df = required_df[['strRequired', 'dexRequired', 'intRequired', 'faiRequired', 'arcRequired']]

# Concatenar required_df con weapons_df
weapons_df = pd.concat([weapons_df, required_df], axis=1)

del required_df

weapons_df.head()
```

Las columnas `attack` y `requiredAttributes` han sido extraidas correctamente, el siguiente paso es continuar con el procesamiento del dataframe `weapons_df`.

```python
weapons_df = weapons_df.drop(['attack', 'requiredAttributes'], axis=1)

# Rellenar valores nulos con 0 en las columnas de atributos requeridos
columns_list = ['strRequired', 'dexRequired', 'intRequired', 'faiRequired', 'arcRequired']
for column in columns_list:
    weapons_df[column] = weapons_df[column].fillna(0)
```

```python
weapons_df.info()
```

Nuevamente se verifica la integridad de los datos del dataframe. Se observa que ya no hay valores nulos, sin embargo las columnas correspondientes a los atributos requeridos están con tipo de dato float, hay que transformarlas a enteros.

```python
weapons_df[['strRequired', 'dexRequired', 'intRequired', 'faiRequired', 'arcRequired']] = weapons_df[['strRequired', 'dexRequired', 'intRequired', 'faiRequired', 'arcRequired']].astype(np.int64)
weapons_df.dtypes
```
